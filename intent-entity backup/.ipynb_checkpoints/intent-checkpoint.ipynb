```python
import pandas as pd

def save_to_csv(text_list, intent, filename='intent1.csv'):
   
    df = pd.DataFrame({'sentences': text_list, 'category': [intent] * len(text_list)})
    
    
    df.to_csv(filename, index=False, encoding='utf-8')
    print(f"Data saved to {filename}")

def read_txt_file(filepath):
    """Reads a text file and returns a list of lines."""
    with open(filepath, 'r', encoding='utf-8') as file:
        return [line.strip() for line in file.readlines() if line.strip()]


input_file = "C.txt"  
text_samples = read_txt_file(input_file)

assigned_intent = "CheckBalance"

save_to_csv(text_samples, assigned_intent)
```

```python
def append_csv(source_file, target_file):
    df = pd.read_csv(source_file)  
    df.to_csv(target_file, mode='a', index=False, header=False)  
    print(f"Data from {source_file} appended to {target_file}")


source_csv = 'N20.csv' 
target_csv = 'BankingIntent.csv' 

append_csv(source_csv, target_csv)
```

```python
import pandas as pd
import re


df2 = pd.read_csv("Bankintent.csv")


def clean_text(text):
    if pd.isnull(text):
        return ""  
    text = re.sub(r'[-]', ' ', text)  
    text = re.sub(r'[^\w\s]', '', text)  # Rem
    text = re.sub(r'["]', '', text)  # Remove double quotes
    text = text.lower().strip()  # Convert to lowercase and strip spaces
    return text

# Apply cleaning only to 'sentences' column
df2['sentences'] = df2['sentences'].apply(clean_text)

# Verify changes
print(df2.head())
```

```python
import pandas as pd

df2 = df2.sample(frac=1, random_state=42).reset_index(drop=True) 

train_size = int(0.75 * len(df2))  
dev_size = len(df2) - train_size  

train_df = df2.iloc[:train_size]
dev_df = df2.iloc[train_size:]

train_df.to_csv("train_data.csv", index=False)
dev_df.to_csv("dev_data.csv", index=False)

print(f"Train Data: {len(train_df)} rows")
print(f"Dev Data: {len(dev_df)} rows")
```

```python
import pandas as pd
import spacy
from spacy.tokens import DocBin
from tqdm import tqdm

df3=pd.read_csv("train_data.csv")

nlp = spacy.blank("en")  
textcat = nlp.add_pipe("textcat")

intents = df3["category"].unique()
for intent in intents:
    textcat.add_label(intent)

db = DocBin()

for _, row in tqdm(df3.iterrows(), total=len(df3)):
    text = row["sentences"].lower()
    intent = row["category"]
    
    doc = nlp.make_doc(text)  
    doc.cats = {label: 1.0 if label == intent else 0.0 for label in intents} 
    db.add(doc)  

db.to_disk("./train.spacy")
```

```python
import pandas as pd
import spacy
from spacy.tokens import DocBin
from tqdm import tqdm


df4=pd.read_csv("dev_data.csv")

nlp = spacy.blank("en")  

textcat = nlp.add_pipe("textcat")

intents = df4["category"].unique()
for intent in intents:
    textcat.add_label(intent)

db = DocBin()

for _, row in tqdm(df4.iterrows(), total=len(df4)):
    text = row["sentences"].lower()
    intent = row["category"]
    
    doc = nlp.make_doc(text)  
    doc.cats = {label: 1.0 if label == intent else 0.0 for label in intents}  
    db.add(doc) 

db.to_disk("./dev.spacy")
```

```python
import spacy

model = spacy.load('IntentP8')  

content = "p on my other account of 50000 rupees"

doc = model(content)

print("Processed tokens:", [token.text for token in doc])


print(f"Text: {content}")

predicted_intent = max(doc.cats, key=doc.cats.get)
print(f"Predicted Intent: {predicted_intent} (Confidence: {doc.cats[predicted_intent]:.4f})")

print("All intent scores:", doc.cats)
```
